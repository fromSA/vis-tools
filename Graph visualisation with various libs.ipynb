```python
from bokeh.io import show, output_notebook
from bokeh.models import GraphRenderer, Oval, StaticLayoutProvider, LabelSet, ColumnDataSource
from bokeh.palettes import Spectral8
from bokeh.plotting import figure

# define nodes and edges
nodes = ["Control Theory", "Emergence", "Interdependence", "Adaptability", 
         "Feedback", "Dynamic Feedback", "System State", "Criticality", 
         "Subsystems", "Network Structure"]
edges = [("Control Theory", "Emergence"), 
         ("Control Theory", "Interdependence"), 
         ("Control Theory", "Adaptability"),
         ("Control Theory", "Feedback"),
         ("Control Theory", "Dynamic Feedback"),
         ("Control Theory", "System State"),
         ("Control Theory", "Criticality"),
         ("Control Theory", "Subsystems"),
         ("Control Theory", "Network Structure"),
         ("Dynamic Feedback", "System State"),
         ("System State", "Criticality"),
         ("Subsystems", "System State"),
         ("Subsystems", "Feedback"),
         ("Interdependence", "Adaptability"),
         ("Interdependence", "Feedback"),
         ("Interdependence", "Subsystems"),
         ("Network Structure", "Interdependence")]

# create a dictionary of node positions
node_positions = {
    "Control Theory": (1, 1),
    "Emergence": (2, 2),
    "Interdependence": (2, 1),
    "Adaptability": (2, 0),
    "Feedback": (3, 2),
    "Dynamic Feedback": (3, 1),
    "System State": (3, 0),
    "Criticality": (4, 0.5),
    "Subsystems": (4, 2),
    "Network Structure": (5, 1)
}

# create a list of node colors
node_colors = [Spectral8[i % 8] for i in range(len(nodes))]

# create a dictionary of node sizes
node_sizes = {
    "Control Theory": 50,
    "Emergence": 40,
    "Interdependence": 40,
    "Adaptability": 40,
    "Feedback": 40,
    "Dynamic Feedback": 40,
    "System State": 40,
    "Criticality": 50,
    "Subsystems": 40,
    "Network Structure": 40
}

# create a dictionary of node labels
node_labels = {
    "Control Theory": "Control Theory",
    "Emergence": "Emergence",
    "Interdependence": "Interdependence",
    "Adaptability": "Adaptability",
    "Feedback": "Feedback",
    "Dynamic Feedback": "Dynamic Feedback",
    "System State": "System State",
    "Criticality": "Criticality",
    "Subsystems": "Subsystems",
    "Network Structure": "Network Structure"
}

# create a ColumnDataSource for nodes
node_data = dict(
    index=nodes,
    color=node_colors,
    size=[node_sizes[node] for node in nodes],
    label=[node_labels[node] for node in nodes]
)
node_source = ColumnDataSource(node_data)

# create a ColumnDataSource for edges
edge_data = dict(
    start=[edge[0] for edge in edges],
    end=[edge[1] for edge in edges]
)
edge_source = ColumnDataSource(edge_data)

# create a graph renderer and set its attributes
graph = GraphRenderer()
graph.node_renderer.data_source = node_source
graph.node_renderer.glyph = Oval(height=0.2, width=0.5, fill_color="color")
graph.edge_renderer.data_source = edge_source
graph.edge_renderer.glyph.line_alpha = 0.8
graph.edge_renderer.glyph.line_width = 2
graph.edge_renderer.glyph.line_color = "#cccccc"

# set the layout provider for the graph
graph_layout = dict(zip(nodes, [node_positions[node] for node in nodes]))
layout_provider = StaticLayoutProvider(graph_layout=graph_layout)
graph.layout_provider = layout_provider

# create labels for nodes
label_source = LabelSet(x='x', y='y', text='label', source=node_source, text_font_size='10pt')

# create a plot and add the graph and labels to it
plot = figure(title="Relationships between Thesis Ideas", x_range=(-0.5, 5.5), y_range=(-0.5, 2.5),
              plot_width=800, plot_height=400, tools="", toolbar_location=None)
plot.renderers.append(graph)
plot.renderers.append(label_source)

# remove axis ticks and labels
plot.axis.visible = False

# display the plot in a notebook
output_notebook()
show(plot)
```
