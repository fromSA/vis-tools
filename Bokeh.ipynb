```python
import bokeh
```

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
```

```python
# show output in the jupyter notebook, otherwise, it will pop-up in a new brower tab.
output_notebook()
```

```python
p = figure(width=400, height=400)

# Data
x = [1, 2, 3, 4, 5]
y = [6, 7, 2, 4, 5]


# add glyphs

p.circle(x , y , size=20, color="navy", alpha=0.5)

p.line(x, y, line_width=2)

# styling

p.title.text = "Trying bokeh"
p.title.text_color = "olive"
p.title.text_font = "times"
p.title.text_font_style = "bold"

p.background_fill_color = "beige"
p.background_fill_alpha = 0.5

p.outline_line_width = 7
p.outline_line_alpha = 0.3
p.outline_line_color = "navy"


# show figure

show(p)
```

```python
def plot():
    p = figure(width=400, height=400)

    # Data
    x = [1, 2, 3, 4, 5]
    y = [6, 7, 2, 4, 5]


    # add glyphs

    p.circle(x , y , size=20, color="navy", alpha=0.5)

    p.line(x, y, line_width=2)

    # styling

    p.title.text = "Trying bokeh"
    p.title.text_color = "olive"
    p.title.text_font = "times"
    p.title.text_font_style = "bold"

    p.background_fill_color = "beige"
    p.background_fill_alpha = 0.5

    p.outline_line_width = 7
    p.outline_line_alpha = 0.3
    p.outline_line_color = "navy"
    
    return p
```

```python
from bokeh.layouts import layout
```

```python
# Layout graphs
plot_1, plot_2, plot_3, plot_4 = plot(), plot(), plot(), plot()
l = layout([[plot_1, plot_2], [plot_3, plot_4]])
show(l)
```
